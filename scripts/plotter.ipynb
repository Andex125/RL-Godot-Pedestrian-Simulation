```python
from pedpy import WalkableArea

curve_c_area = WalkableArea(
    # Complete area (outer boundary)
    [
        (10, -10),
        (10, 10),
        (-10, 10),
        (-10, -10),
    ],
    obstacles=[
        # bottom barrier
        [
            (9.999, -6.5),
            (-7, -6.5),
            (-7, 6.5),
            (9.999, 6.5),
            (9.999, 5.5),
            (-6, 5.5),
            (-6, -5.5),
            (9.999, -5.5),
        ],
        
    ],
)
```

```python
from pedpy import plot_measurement_setup
import matplotlib.pyplot as plt

plot_measurement_setup(
    walkable_area=curve_c_area,
    traj_alpha=0.5,
    traj_width=1,
    ma_line_width=2,
    ma_alpha=0.5,
).set_aspect("equal")
plt.show()
```
